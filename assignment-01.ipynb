```python
#importing important liabraries
import seaborn as sns 
import matplotlib.pyplot as plt 
```

**UNIVARIATE DATA ANALYSIS**


```python
#defining a univaraite function for performing anaylsis of every column
def ua(column):
    print(f"\nUnivariate Analysis for {column}:")
    print(df[column].describe())  # Descriptive statistics
    
    plt.figure(figsize=(12, 6))

    
    # Check if the column is numerical or categorical
    if df[column].dtype != 'object':  # Numerical column
        # Histogram
        plt.subplot(1, 2, 1)
        sns.histplot(df[column], kde=True, color='blue', bins=20)
        plt.title(f'Histogram of {column}')
        plt.xlabel(column)
        plt.ylabel('Frequency')
        
        # Box Plot
        plt.subplot(1, 2, 2)
        sns.boxplot(x=df[column], color='orange')
        plt.title(f'Box Plot of {column}')
    
    else:  # Categorical column
        # Bar Plot
        sns.countplot(x=df[column], palette='viridis')
        plt.title(f'Bar Plot of {column}')
        plt.xlabel(column)
        plt.ylabel('Count')
    
    plt.tight_layout()
    plt.show()

# Perform univariate analysis for all columns
for column in df.columns:
    ua(column)
```
